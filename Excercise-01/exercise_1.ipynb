# Question 09

```python
def hyperbolid_estimate(s):
    n = 0
    a = s
    while a >= 100:
        n+=1
        a = a / (100)
    return ((-190/(a + 20))+10)*10**n
```

```python
def newton_raphson(s, precision):
    s0 = hyperbolid_estimate(s)
    while True:
        sqrt = 0.5 * (s0 + (s / s0))
        if (abs(sqrt - s0)<precision):
            break
        s0 = sqrt
    return sqrt
```

```python
precision = 10**(-5)
print(newton_raphson(64,precision))
print(newton_raphson(75,precision))
print(newton_raphson(100,precision))
print(newton_raphson(1600,precision))
```
